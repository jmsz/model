```python

source_activity = 200.0 # gammas per s
#inner_radius = 1 # m
#detector_x = 0.1 # m
#detector_y = 0.1 # m
#detector_area = detector_x * detector_y # m^2
#solid_angle = detector_area / (4 * np.pi * inner_radius ** 2)

solid_angle= 0.5

incidient_gamma_energy = 500 # keV
incident_gamma = source_activity * solid_angle

# at 500 keV
mu_rho = 8.212E-02 #at 500 keV
pe_mu_rho = 3.541E-03 #at 500 keV, just photo electric
# mu_en_rho = 2.930E-02
density_ge = 5.323 # g/cm^3

# canberra BEGe detector p-type, 5% efficiecncy at 500 keV

number_of_layers_list = np.linspace(1, 3, 2) #cm
thickness_list = np.linspace(0, 5, 5) # cm

number_of_layers = 2

#detector_area = solid_angle * (4 * np.pi * inner_radius_1 ** 2)
#print (detector_area)

total_stopped = 0
efficiency_list = []
incident_gamma_list = []
outgoing_gamma_list = []
total_stopped_list = []

for thickness in thickness_list:
    absorbed_gamma = 0
    incident_gamma_list.append(incident_gamma)
    outgoing_gamma = incident_gamma * np.exp(-mu_rho * density_ge * thickness)
    absorbed_gamma = incident_gamma - incident_gamma * np.exp(-pe_mu_rho * density_ge * thickness)
    outgoing_gamma_avg_energy = 500 # TODO

    outgoing_gamma_list.append(outgoing_gamma)

    #incident_gamma = outgoing_gamma
    #outgoing_gamma = 0
    #total_stopped = total_stopped + absorbed_gamma
        
    efficiency_list.append(absorbed_gamma / incident_gamma)
    total_stopped_list.append(absorbed_gamma)

print("thickness")
print(thickness_list)
print("efficiency")
print(efficiency_list)
print("incident_gamma_list")
print(incident_gamma_list)
print("outgoing_gamma_list")
print(outgoing_gamma_list)
print("total_stopped_list")
print(total_stopped_list)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(thickness_list, efficiency_list, linestyle='-', marker = 'o', color='blue')
plt.legend(loc='upper right', ncol=2)
plt.title('thickness vs efficiency')
plt.xlabel('total thickness (all layers) (cm)')
plt.ylabel('Efficiency (# photoelectric absorptions/ # incident gamma rays)')
plt.show()
```
